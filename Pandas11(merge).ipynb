# Pandas Merge

```python
df =pd.read_csv('C:\\Users\\Nikhil Gupta\\ML_Pandas\\student_results.csv')
df
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df1
```

```python
df2 =pd.DataFrame({'A':[1,2,3,4],'C':[51,61,71,81]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='A')
```

```python
pd.merge(df2,df1,on='A')
```

```python
df3 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df4 =pd.DataFrame({'A':[1,2,3,5],'C':[51,61,71,81]})
pd.merge(df3,df4,on ='A')
```

```python
pd.merge(df3,df4,on='A',how ='left')
```

```python
pd.merge(df3,df4,on='A',how ='right')
```

```python
pd.merge(df3,df4,on='A',how ='outer')
```

```python
pd.merge(df3,df4,on='A',how ='outer',indicator = True)
```

```python
pd.merge(df3,df4,how ='outer',indicator =True)
```

```python
pd.merge(df3,df4,on='A',how ='right',indicator =True)
```

```python
df5 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df6 =pd.DataFrame({'A':[1,2,3,4],'B':[15,16,17,18]})
pd.merge(df5,df6)
```

```python
pd.merge(df5,df6,left_index =True,right_index =True)
```

```python
pd.merge(df5,df6,left_index =True,right_index =True,
        suffixes =('name','ID'))
```
