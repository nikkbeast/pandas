# Pandas Concat

```python
import pandas as pd
```

```python
s1 =pd.Series([1,2,3,4])
s2 =pd.Series([11,12,13,14])
pd.concat([s1,s2])
```

```python
s1 =pd.Series([1,2,3,4])
s2 =pd.Series([11,12,13,14])
s3 =pd.Series([11,12,13,14])
pd.concat([s1,s2,s3])
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2,3,5],'B':[51,61,71,81]})
pd.concat([df1,df2])
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2,3,4],'B':[51,61,71,81]})
pd.concat([df1,df2])
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2,3,5],'B':[51,61,71,81]})
pd.concat([df1,df2],axis =1)
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2],'C':[51,61]})
pd.concat([df1,df2],axis =1)
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2],'C':[51,61]})
pd.concat([df1,df2],axis =1,join ='outer')
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'C':[1,2],'D':[51,61]})
pd.concat([df1,df2],axis =1,join='inner')
```

```python
df1 =pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
df2 =pd.DataFrame({'A':[1,2,3,5],'B':[51,61,71,81]})
pd.concat([df1,df2],axis =1,keys =['c1','c2'])
```
