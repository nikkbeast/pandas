```python
import pandas as pd
```

# Melt

```python
df1 =pd.DataFrame({'Roll_No.':[1,2,3,4,5,6],'eng':[5,6,7,8,9,12],'maths':[12,14,15,17,13,19]})
df1
```

```python
pd.melt(df1)
```

```python
pd.melt(df1,id_vars =["Roll_No."])
```

```python
pd.melt(df1,id_vars =["Roll_No."],var_name = 'Subject')
```

```python
pd.melt(df1,id_vars =["Roll_No."],var_name = 'Subject',value_name ="Marks")
```
