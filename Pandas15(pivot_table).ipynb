```python
import pandas as pd
```

# Pivot()

```python
df =pd.read_csv('C:\\Users\\Nikhil Gupta\\ML_Pandas\\Movie_Ratings.csv')
df
```

```python
pd.pivot_table(df,index ="Genre")
```

```python
df.pivot_table(index = "Genre")
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year')
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year' , aggfunc = 'count')
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year' , aggfunc = 'sum')
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year' , aggfunc = 'max')
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year' , aggfunc = 'count',fill_value = 0)
```

```python
pd.pivot_table(df,index ="Genre",columns = 'Year' ,fill_value = 0,margins =True)
```
